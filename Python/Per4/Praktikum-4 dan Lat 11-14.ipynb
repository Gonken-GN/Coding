## Latihan No 11
Patut diduga rata-rata tinggi badan dan berat badan bergantung pada gender. Oleh karena itu akan lebih akurat jika data yang kosong diisi dengan median sesuai atribut Sex. Buat kode untuk melihat apakah dugaan ini benar, dan buat kode untuk mengisi berdasarkan median atribut Sex. 

```python
df2.isnull().sum()

#Cek atribut yang masih memiliki data kosong/null
```

```python
med_height_m = df2[df2.Sex == "M"]["Height"].median()
med_weight_m = df2[df2.Sex == "M"]["Weight"].median()
med_height_m

# Hitung median tinggi dan berat berdasarkan jenis kelamin Laki-laki
```

```python
med_height_f = df2[df2.Sex == "F"]['Height'].median()
med_weight_f = df2[df2.Sex == "F"]["Weight"].median()
med_height_f

# Hitung median tinggi dan berat berdasarkan jenis kelamin Perempuan
```

```python
df2.loc[df2.Sex=="M","Height"] = df2.loc[df2.Sex=="M","Height"].fillna(med_height_m)
df2.loc[df2.Sex=="F","Height"] = df2.loc[df2.Sex=="F","Height"].fillna(med_height_f)

# Masukkan data median tinggi ke dalam dataframe
```

```python
df2.loc[df2.Sex=="M","Weight"] = df2.loc[df2.Sex=="M","Weight"].fillna(med_weight_m)
df2.loc[df2.Sex=="F","Weight"] = df2.loc[df2.Sex=="F","Weight"].fillna(med_weight_f)

# Masukkan data median tinggi ke dalam dataframe
```

```python
df2.Height.isnull().sum()

# periksa jumlah data tinggi yang masih null
```

## Latihan no 13
Atribut apa saja yang dapat ditambahkan berdasarkan atribut yang ada?  (tentu yang kira-kira relevan dengan perolehan medali). Tulis sebanyak-banyaknya dan buat code-nya.

```python
df2['Medal'].fillna("Non-Medal", inplace=True)
df2['Medal'].head()

#Mengubah data yang peserta yang tidak mendapat medal menjadi "Non-Medal"
```

```python
df2['Podium'] = df2.apply(lambda row: "Juara" if (row.Medal == "Gold") | (row.Medal == "Silver") | (row.Medal == "Bronze") else "Peserta", axis=1)
df2['Podium'] = df2["Podium"].astype("category")
df2[['Medal','Podium']].head(20)

# Menambah Kategori Podium, untuk yang mendapat medal gold, silver dan bronze menjadi kategori "Juara" dan
# yang tidak mendapat medal masuk kategori "Peserta"
```
